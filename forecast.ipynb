```python
#from __future__ import print_function
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing
from sklearn.cross_validation import cross_val_score # K折交叉验证模块
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_predict
import numpy as np
import time

data = pd.DataFrame(pd.read_csv('wanfupassenger4v2.csv',header=0))
x = data.ix[:,2:]
x = preprocessing.scale(x)
y = data['y']
y = y.reshape(-1)
model = LinearRegression()
predicted = cross_val_predict(model, x, y, cv=10)
#print(predicted)
fig, ax = plt.subplots()
ax.scatter(y, predicted, edgecolors=(0, 0, 0))
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
#ax.set_xlabel('Measured')
#ax.set_ylabel('Predicted')
plt.show()

#print(y)
#X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
score_mean = []
for i in range(0, 1000):
    model = LinearRegression()
    data_train,data_test = train_test_split(data, test_size=0.3)
    X_train = data_train.ix[:,2:]
    X_train = preprocessing.scale(X_train)
    y_train = data_train['y']
    y_train = y_train.reshape(-1)
    X_test = data_test.ix[:,2:]
    X_test = preprocessing.scale(X_test)
    y_test = data_test['y']
    y_test = y_test.reshape(-1)
    #print(y_train)

    model.fit(X_train,y_train)
    forecast = model.predict(X_test)
    #print(model.get_params)
    #得分
    y_test_score = model.score(X_test,y_test)
    #print(y_test_score)
    score_mean.append(y_test_score)
sum = 0
for i in score_mean:
    sum = i+sum
print(sum)
print(sum/len(score_mean))
#print(forecast)
#print(y_test)





data_test.insert(2,'Forecast',forecast)
out_put = data_test[['date','y','Forecast']]
print(out_put.shape)
percent=[]
out_put = out_put.reset_index(drop=True)  
#print(out_put.shape[0])
print(out_put.shape[0])
print(out_put.iat[0,0])
for i in range(out_put.shape[0]):
    #print(out_put[1])
    percent.append(abs((out_put.iat[i,2]-out_put.iat[i,1])/out_put.iat[i,1]))
#print(percent)
data_test.insert(3,'Percent',percent)
out_put = data_test[['date','y','Forecast','Percent']]
out_put = out_put.sort_values(by='Percent')  
print(out_put)
#np.insert(data,2,values=y_test,axis=1)
#np.c_[X_test,y_test]
#np.savetxt('2.txt',X_test[[0,1,17]])
#output = data_test['data','y','Forecast']
#output = data[[0,1,2]]
#output = pd.DataFrame(columns=['date','Real','Forecast'])
#output.insert(0,'date',X_test['date'])
#output.insert(1,'Real',y_test)
#output.insert(2,'Forecast',forecast)

#print(out_put)
out_put.to_csv(str(time.strftime("%Y-%m-%d %H:%M:%S", time.localtime())).replace(':','.')+'.csv',index=False,sep=',')  
```
